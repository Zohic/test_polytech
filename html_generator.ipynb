```
class Tag:
    def __init__(self):
        self.tagStr = "tagName"
        self.innerHTML = ""
        self.attributes = ""
        self.container = []

    def Generate(self):
        ostr = "<"+self.tagStr+" "+self.attributes+">"+self.innerHTML+" \n"
        for tag in self.container:
            ostr = ostr+tag.Generate()
        ostr = ostr+"</"+self.tagStr+"> \n"
        return ostr

    def GenerateFile(self, path):
        file = open(path, "w")
        file.write(self.Generate())
        file.close

    pass
#--------------------------------------------------------

class Img(Tag):
    def __init__(self, attr):
        Tag.__init__(self)
        self.tagStr = "img"
        self.attributes = attr
    pass

class Br(Tag):
    def __init__(self):
        Tag.__init__(self)
        self.tagStr = "br"

class Script(Tag):
    def __init__(self, text, attr):
        Tag.__init__(self)
        self.tagStr = "script"
        self.innerHTML = text
        self.attributes = attr
    pass

class Style(Tag):
    def __init__(self, text, attr):
        Tag.__init__(self)
        self.tagStr = "style"
        self.innerHTML = text
        self.attributes = attr
    pass

class tagA(Tag):
    def __init__(self,text, attr):
        Tag.__init__(self)
        self.tagStr = "a"
        self.innerHTML = text
        self.attributes = attr
    pass

class Title(Tag):
    def __init__(self,text):
        Tag.__init__(self)
        self.tagStr = "title"
        self.innerHTML = text
    pass


class Html(Tag):
    def __init__(self, data):
        Tag.__init__(self)
        self.container = data
        self.tagStr = "html"
    pass

class Div(Tag):
    def __init__(self, data, attr):
        Tag.__init__(self)
        self.tagStr = "div"
        self.container = data
        self.attributes = attr

class Head(Tag):
    def __init__(self, data):
        Tag.__init__(self)
        self.tagStr = "head"
        self.container = data
    pass

class Body(Tag):
    def __init__(self, data,attr):
        Tag.__init__(self)
        self.tagStr = "body"
        self.container = data
        self.attributes = attr
    pass



oo = Img("image.com")

page = Html([
    Head([Title("THE TITLE")]),
    Script("alert('hello')","type='text/javascript'"),
    Style("a { font-size: 150%; }","type='text/css'"),
    Body([
        Div([
            tagA("click here","href='https://www.google.com'"),
            Br(),
            Img("src='https://i.playground.ru/i/pix/482939/image.jpg' width='100px' height='100px'")
            ],"")
        
        ],"")
    ])

#page.GenerateFile("wow.html")
print(page.Generate())
```
